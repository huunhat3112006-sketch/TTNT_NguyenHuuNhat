```python
import numpy as np
a = np.array([3,9,1,4])
b=np.array([2,7,4,3,2,8])
min_len = min(len(a),len(b))
if(len(a)>len(b)):
  c = np.copy(a)
else:
  c = np.copy(b)
c[:min_len] = np.minimum(a[:min_len],b[:min_len])
if len(a)>len(b):
  phan_du = a[min_len:]
elif len(b)>len(a):
  phan_du=b[min_len:]
else:
  phan_du=np.array([])
print(f"Mảng đã trộn:{c}")
print(f"Mảng dư sau khi trộn:{phan_du}")
```

```python
import numpy as np
n=int(input("Nhập kích thước ma trận vuông n:"))
a = np.random.randint(-20,20,size=(n,n))
print(a)
def tong_tam_giac_tren(a):
  n=len(a)
  dong,cot=np.indices((n,n))
  tam_giac_tren = (dong+cot)<=(n-1)
  return np.sum(a[tam_giac_tren])
def chuyen_tri_tuyet_doi(a):
  return np.abs(a)
def thay_phan_tu_chan(a,x):
  ket_qua=np.where(a%2==0,x,a)
  return ket_qua
def kiem_tra_chan(a):
  return np.all(a%2==0)
def kiem_tra_doi_xung(a):
  return np.array_equal(a,a.T)
def kiem_tra_duong_cheo_chinh(a):
  cheo_chinh=np.diagonal(a)
  return np.all(np.diff(cheo_chinh)>0)
def xuat_tam_giac_duoi(a):
  n=len(a)
  dong,cot=np.indices((n,n))
  tam_giac_duoi=(dong+cot)>=(n-1)
  return a[tam_giac_duoi]
def kiem_tra_duong_cheo_phu(a):
  ma_tran_nguoc=np.fliplr(a)
  cheo_phu=np.diagonal(ma_tran_nguoc)
  return np.all(np.diff(cheo_phu)<0)
print(f"Tổng các phần tử tam giác trên:{tong_tam_giac_tren(a)}")
print(f"Ma trận sau khi trị tuyệt đối:{chuyen_tri_tuyet_doi(a)}")
x=int(input("Nhập giá trị thay thế:"))
print(f"Ma trận sau khi thay thế: {thay_phan_tu_chan(a,x)}")
if kiem_tra_chan(a):
  print("Ma trận là ma trận chẵn")
else:
  print("Ma trận không phải là ma trận chẵn")
if kiem_tra_doi_xung(a):
  print("Ma trận là ma trận đối xứng")
else:
  print("Ma trận không phải là ma trận đối xứng")
if kiem_tra_duong_cheo_chinh(a):
  print("Ma trận là ma trận có đường chéo chính tăng dần")
else:
  print("Ma trận không phải là ma trận có đường chéo chính tăng dần")
print(f"Tam giac dưới của ma trận{xuat_tam_giac_duoi(a)}")
if kiem_tra_duong_cheo_phu(a):
  print("Ma trận là ma trận có đường chéo phụ giảm dần")
else:
  print("Ma trận là ma trận ko có đường chéo phụ giảm dần")
```
